# 冒泡排序

```python
lis = [56,12,1,8,354,10,100,34,56,7,23,456,234,-58]

def bubbleSort(ls):
    for i in range(len(ls)):
        for j in range(len(ls)):
            if ls[j] > ls[i]:
                ls[i], ls[j] = ls[j], ls[i]
    return ls
```

```python
bubbleSort(lis)
```

# 计算x的n次方

```python
def power(x,n):
    s = 1
    while n > 0:
        s = s * x
        n = n - 1
    return s
```

```python
power(2,3)
```

# 计算阶乘

```python
def factorial(n):
    if n < 0:
        print("Input error") 
    elif n == 0:
        return 1 
    else:
        result = n
        for i in range(1,n,1):
            result *= i
        return result
```

```python
factorial(4)
```

```python
factorial(-1)
```

# 输出某个路径下的所有文件和文件夹的路径

```python
import os
def print_dir():
    filepath = input("请输入一个路径：")
    if filepath == "":
        print("请输入正确的路径")
    else:
        for i in os.listdir(filepath):
            print(os.path.join(filepath,i))
```

```python
print_dir()
```

# 输出某个路径及其子目录下的所有文件路径

```python
import os
def show_dir(filepath):
    for i in os.listdir(filepath):
        path = (os.path.join(filepath,i))
        print(path)
        if os.path.isdir(path):
            show_dir(path)  #递归方法
```

```python
filepath = r'/root/workspace/'
show_dir(filepath)
```

# 输出某个路径及其子目录下所有以.html为后缀的文件

```python
import os
def print_dir_condition(filepath):
    for i in os.listdir(filepath):
        path = os.path.join(filepath,i)
        if os.path.isdir(path):
            print_dir(path)
        if path.endswith('.html'):
            print(path)    
```

```python
filepath = r'/root/workspace/'
print_dir_condition(filepath)
```

# 随机生成验证码的两种方式

```python
import random
list1 = []
for i in range(65,91):
    list1.append(chr(i))
for j in range(97,123):
    list1.append(chr(j))
for k in range(48,58):
    list1.append(chr(k))
    
ma = random.sample(list1,6)
print(ma)
ma = ''.join(ma)  #将列表转化为字符串
print(ma)
```

# 斐波那契数列

```python
nterms = int(input())
n1 = 0
n2 = 1
if nterms <= 0:
    print("Please input a positive number:")
elif nterms == 1:
    print(n1)
else:
    print(n1,",",n2,end=" , ")
    count = 2
    while count < nterms:
        nth = n1 + n2
        print(n1+n2,end=' , ')
        #更新值
        n1 = n2
        n2 = nth
        count += 1
```

# 最大公约数

```python
def hcf(x,y):
    '''返回两个数的最大公约数'''
    if x > y:
        smaller = y
    else:
        smaller = x
    
    for i in range(1,smaller+1):
        if x%i==0 and y%i==0:
            hcf = i
    return hcf
```

```python
hcf(18,6)
```

# 最小公倍数

```python
def lcm(x,y):
    if x > y:
        greater = x
    else:
        greater = y
    for i in range(greater,x*y+1):
        if i%x==0 and i%y==0:
            lcm = i
            break
    return lcm
```

```python
lcm(3,12)
```
